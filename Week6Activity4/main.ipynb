```python
factorial = lambda n: n if n<=1 else n*factorial(n-1)

print(factorial(5))  # 120
```

## small and anonymous fucntion: lambda

```python
data = []
for i in range(5):
    data.append(lambda a, i=i*2: i*a)
print(data)
print(data[0](10))
print(data[1](10))
print(data[2](10))
print(data[3](10))
print(data[4](10))
```

```python
data = ["a10", "a2","b1","b3","c2"]
print(sorted(data))
print(sorted(data, key=lambda x:(x[0],int(x[1:]))))

data1 = ["a10e", "a2","b1","b3","c2"]
print(sorted(data1))
print(sorted(data1, key=lambda x:(x[0],int(x[1:]))))
```
